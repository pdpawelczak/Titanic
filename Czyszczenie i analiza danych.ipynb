```python
# Oczyszczanie danych zaczniemy od usunięcia zbędnej kolumny ship
file = file.drop(columns=['ship'])
```

```python
# PassengerId zawiera duplikaty dlatego w kolejnym kroku je usuniemy.
```

```python
file = file.drop_duplicates('PassengerId', ignore_index=True)
```

```python
len(file) == file['PassengerId'].nunique()
```

```python
# Wartość true oznacza, że nie ma wierszy, które posiadają to samo id.
```

```python
# Następnie pozbędziemy się wartości Na, Nan
file['Embarked'] = file['Embarked'].replace(['nan'],'NaN')
file = file.fillna(0)
```

```python
# W kolejnym etapie usunięte zostaną literówki w kolumnie Sex or Embarked
```

```python
file['Sex'] = file['Sex'].replace(['femmale'],'female')
file['Sex'] = file['Sex'].replace(['malef'],'male')
file['Sex'] = file['Sex'].replace(['mal'],'male')

file['Sex'] = file['Sex'].replace(['fem'],'female')
```

```python
pd.concat([file['Sex'],file['Embarked']]).unique()
```

```python
# Analogicznie postępujemy w przypadku Embarked
file['Embarked'] = file['Embarked'].replace(['So'],'S')
file['Embarked'] = file['Embarked'].replace(['Co'],'C')
file['Embarked'] = file['Embarked'].replace(['Qe'],'Q')
pd.concat([file['Sex'],file['Embarked']]).unique()
```

```python
# Następnie usuniemy wartości w kolumnie Age poniżej 1 oraz więcej niż 120. Najpierw zmienimy wartości w Age na typ flaot. 
# Aby tego dokonać najpierw trzeba zmienić przecinki na kropki.
file['Age']=file['Age'].str.replace(',','.')
file['Age'] = file['Age'].astype(float).round(1)
file = file[file['Age'] > 0]
file = file[file['Age'] < 121]


```

```python
file['Age'].unique()
```

```python
# Następnie oczyszczone zostaną wartości z kolumny Fare. Przedstawia ona ceny, które powinny zostac zaokrąglone do 2 miejsc po
# przecinku.Podobnie jak w przypadku wieku trzeba zamienić przecinki na kropki aby dokonac konwersji do floata. Należy także
# wykluczyć wartości ujemne. Należy także podmienić jedną błędną komórkę z 15.9a na 15.9
file['Fare']=file['Fare'].str.replace('15,9a','15,9')
file['Fare']=file['Fare'].str.replace(',','.')
file['Fare'] = file['Fare'].astype(float).round(2)
file = file[file['Fare'] > 0]
```
