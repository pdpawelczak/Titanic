```python
# Oczyszczanie danych zaczniemy od usunicia zbdnej kolumny ship
file = file.drop(columns=['ship'])
```

```python
# PassengerId zawiera duplikaty dlatego w kolejnym kroku je usuniemy.
```

```python
file = file.drop_duplicates('PassengerId', ignore_index=True)
```

```python
len(file) == file['PassengerId'].nunique()
```

```python
# Warto true oznacza, e nie ma wierszy, ktre posiadaj to samo id.
```

```python
# Nastpnie pozbdziemy si wartoci Na, Nan
file['Embarked'] = file['Embarked'].replace(['nan'],'NaN')
file = file.fillna(0)
```

```python
# W kolejnym etapie usunite zostan literwki w kolumnie Sex or Embarked
```

```python
file['Sex'] = file['Sex'].replace(['femmale'],'female')
file['Sex'] = file['Sex'].replace(['malef'],'male')
file['Sex'] = file['Sex'].replace(['mal'],'male')

file['Sex'] = file['Sex'].replace(['fem'],'female')
```

```python
pd.concat([file['Sex'],file['Embarked']]).unique()
```

```python
# Analogicznie postpujemy w przypadku Embarked
file['Embarked'] = file['Embarked'].replace(['So'],'S')
file['Embarked'] = file['Embarked'].replace(['Co'],'C')
file['Embarked'] = file['Embarked'].replace(['Qe'],'Q')
pd.concat([file['Sex'],file['Embarked']]).unique()
```

```python
# Nastpnie usuniemy wartoci w kolumnie Age poniej 1 oraz wicej ni 120. Najpierw zmienimy wartoci w Age na typ flaot. 
# Aby tego dokona najpierw trzeba zmieni przecinki na kropki.
file['Age']=file['Age'].str.replace(',','.')
file['Age'] = file['Age'].astype(float).round(1)
file = file[file['Age'] > 0]
file = file[file['Age'] < 121]


```

```python
file['Age'].unique()
```

```python
# Nastpnie oczyszczone zostan wartoci z kolumny Fare. Przedstawia ona ceny, ktre powinny zostac zaokrglone do 2 miejsc po
# przecinku.Podobnie jak w przypadku wieku trzeba zamieni przecinki na kropki aby dokonac konwersji do floata. Naley take
# wykluczy wartoci ujemne. Naley take podmieni jedn bdn komrk z 15.9a na 15.9
file['Fare']=file['Fare'].str.replace('15,9a','15,9')
file['Fare']=file['Fare'].str.replace(',','.')
file['Fare'] = file['Fare'].astype(float).round(2)
file = file[file['Fare'] > 0]
```
